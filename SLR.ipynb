```python
#DATA

# Advertasing
x = [23,26,30,34,43,48,52,57,58] 

#Sales
y = [651,762,856,1063,1190,1298,1421,1440,1518]

import numpy as np
import matplotlib.pyplot as plt

n = len(x)
x = np.array(x)
y = np.array(y)

sumx = sum(x)
sumy = sum(y)
sumx2 = sum(x*x)
sumy2 = sum(y*y)
sumxy = sum(x*y)
promx = sumx/n
promy = sumy/n

m =  (sumx*sumy - n*sumxy)/(sumx**2 - n*sumx2)
b =  promy - m*promx

print("m :",m)
print("b :" ,b)
```

```python
plt.plot(x, y, 'o' ,label="Datos")
plt.plot(x, m*x + b,label="Ajuste")
plt.xlabel('Advertasing')
plt.ylabel('Sales')
plt.grid()
plt.legend()
plt.show()
```

```python
sigmax = np.sqrt(sumx2/n - promx**2)
sigmay = np.sqrt(sumy2/n - promy**2)
sigmaxy = sumxy/n - promx*promy

rsquared = (sigmaxy/(sigmax*sigmay))**2

print("R Cuadrada: ",rsquared)
```
